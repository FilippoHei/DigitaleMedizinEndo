```python
# For information on the conversion of the labeled dataset to  the coco format visit:
# https://github.com/wkentaro/labelme/tree/master/examples/instance_segmentation

from pycocotools.coco import COCO
import numpy as np
import skimage.io as io
import random
import os
import cv2
from tensorflow.keras.preprocessing.image import ImageDataGenerator


dataDir = 'dataset_coco/Annotaions'
dataType = 'val'
annFile = 'dataset_coco/annotations.json'.format(dataDir,dataType)

# Initialize the COCO api for instance annotations
coco=COCO(annFile)

# Load the categories in a variable
catIDs = coco.getCatIds()
cats = coco.loadCats(catIDs)

annIDs = coco.getCatIds()
anns = coco.loadCats(annIDs)

print(anns)
```

```python
# Now display the number of images containing the given class (label), for each class

filterClasses = ['Nadelhalter']

# Fetch class IDs only corresponding to the filterClasses
catIds = coco.getCatIds(catNms=filterClasses) 
# Get all images containing the above Category IDs
imgIds = coco.getImgIds(catIds=catIds)
print("Number of images containing Nadelhalter:", len(imgIds))

filterClasses = ['Atraum. Pinzette']
catIds = coco.getCatIds(catNms=filterClasses) 
imgIds = coco.getImgIds(catIds=catIds)
print("Number of images containing Atraum. Pinzette:", len(imgIds))

filterClasses = ['Knotenschieber']
catIds = coco.getCatIds(catNms=filterClasses) 
imgIds = coco.getImgIds(catIds=catIds)
print("Number of images containing Knotenschieber:", len(imgIds))

filterClasses = ['Nervhaken']
catIds = coco.getCatIds(catNms=filterClasses) 
imgIds = coco.getImgIds(catIds=catIds)
print("Number of images containing Nervhaken:", len(imgIds))

filterClasses = ['Klappenschere']
catIds = coco.getCatIds(catNms=filterClasses) 
imgIds = coco.getImgIds(catIds=catIds)
print("Number of images containing Klappenschere:", len(imgIds))

filterClasses = ['Artefakt']
catIds = coco.getCatIds(catNms=filterClasses) 
imgIds = coco.getImgIds(catIds=catIds)
print("Number of images containing Artefakt:", len(imgIds))


# display the number of total labels in the dataset

classes = ['Nadelhalter', 'Atraum. Pinzette', 'Knotenschieber', 'Nervhaken', 'Klappenschere']

images = []
for className in classes:
        # get all images containing given class
        catIds = coco.getCatIds(catNms=className)
        imgIds = coco.getImgIds(catIds=catIds)
        images += coco.loadImgs(imgIds)
        
dataset_size = len(images)

print("Number of total labels:", dataset_size)


# unique images are the total images with at least one label

unique_images = []
for i in range(len(images)):
    if images[i] not in unique_images:
        unique_images.append(images[i])

dataset_size_unique = len(unique_images)


# number of images with 2 labels = total labels - unique images with at least one label (if there are max. 2 label per image)

print("Number of images containing more than one label:", dataset_size - dataset_size_unique)
```

```python
# now you need the json dataset from the labelme output
import json
common_json_dir = 'json_dataset'
```

```python
# display number of images containing given shape, e.g. rectangle
searchTerm = "rectangle"
rectangles = 0
for root, dirs, files in os.walk(common_json_dir):
    for file in files:
        j_file = open(os.path.join(root,file), 'r')
        data = json.load(j_file)

        for element in data["shapes"]:
            if element["shape_type"] == searchTerm:
                rectangles += 1
print(rectangles)
```
